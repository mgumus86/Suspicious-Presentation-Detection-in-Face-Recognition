```python
def extract_face(img, precision, fn):
    try:
        face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, precision)
        for (x,y,w,h) in faces:
            img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
            roi_color = img[y:y+h, x:x+w]
            return roi_color
    except Exception as ex:
        print('ex', fn)
    return None
```

```python
def extract_frames(fdir, fn, t='n', s='train', no_samples=1):
    vidcap = cv2.VideoCapture(os.path.join(fdir,fn))
    success,image = vidcap.read()
    count = 0
    precision = 10
    while True:   
        res = extract_face(image, precision, fn)
        if res is not None:
            for c in range(no_samples):
                lpath = f'{s}-{t}//{fn.split(".")[0]}_{c}.jpg' 
                cv2.imwrite(lpath, cv2.resize(res, (150,150)))  
            break
#         if res:
#             fs.put(lpath, f'cda-project/images/{s.title()}/{t}/{fn.split(".")[0]}_{count}.jpg')
        success,image = vidcap.read()    
        count = count + 1
        
        # Try multiple precisions and give up if no face
        if count > 40 and precision == 1:
            print(fn, t, s, precision, 'no face')
            break
        elif count > 40:
            print(fn, t, s, precision)
            vidcap = cv2.VideoCapture(os.path.join(fdir,fn))
            success,image = vidcap.read()
            precision = precision - 1
            count = 0
```
